```python
# バブルソート
# 隣り合うふたつの要素の値を比較して条件に応じた交換を行う整列アルゴリズム

list = [6, 15, 4, 2, 8, 5, 11, 9, 7, 13]

for i in range(len(list)):
    for j in range(len(list) - i - 1): # ソート済み以外の部分でループする
        if list[j] > list[j + 1]: # 前の方が大きい時
            list[j], list[j + 1] = list[j + 1], list[j]
            print(list)            
```

```python
# 交換が起きなかった際に処理を打ち切り高速化
list = [6, 15, 4, 2, 8, 5, 11, 9, 7, 13]
change = True

for i in range(len(list)):
    print(change)
    if not change:
        break
    change = False
    for j in range(len(list) - i - 1): # ソート済み以外の部分でループする
        if list[j] > list[j + 1]: # 前の方が大きい時
            list[j], list[j + 1] = list[j + 1], list[j]
            change = True
            print(list)
            
```
