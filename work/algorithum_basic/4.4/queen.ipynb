```python
# 8クイーン問題
#
# 問題:
# チェスの8×8のボード上に、8つのクイーンを互いに攻撃し合わないように配置する方法を見つけなさい。
# クイーンは、縦、横、および斜めのいずれの方向にも移動できます。
#
# 条件:
# 1. 1列に1つのクイーンを置くこと。
# 2. 各クイーンは、他のクイーンが配置されている同じ行、列、または斜めのライン上に位置してはならない。
#
# 入力:
# チェスボードは空でスタートする（クイーンはまだ置かれていない）。
#
# 出力:
# クイーンを配置したチェスボードの配置を示す。1つ以上の解がある場合、それぞれの解を出力すること。
#
# 例:
#
# 以下は、1つの解の例です（Qはクイーンの位置を示します）:
#　 列
# 行. . Q . . . . .
# 行. . . . Q . . .
# 行. . . . . . Q .
# 行. Q . . . . . .
# 行. . . Q . . . .
# 行. . . . . Q . .
# 行Q . . . . . . .
# 行. . . . . . . Q
# 列がindex, 行をvalue
# 各行には1つのクイーンが含まれ、どのクイーンも他のクイーンと同じ行、列、または斜めに位置していないことが確認できます。
```

```python
# クイーンの数
# 縦軸をy 横軸をxとする
N = 8

# 斜めチェック x:配置するクイーンの行の位置 col:すでに配置されたクイーンの行位置リスト
def check(x, col):
    # 配置済みの行を逆順で調べる
    for i , row in enumerate(reversed(col)):
        if (x + i + 1 == row) or (x - i - 1 == row):
            return False
    return True

# indexが列でvalueが行となる
def search(col):
    if len(col) == N:
        print(col)
        return
    for i in range(N):
        if i not in col: # 同じ行(x)にいないこと
            #  斜めチェック
            if check(i, col):
                col.append(i)
                search(col)
                col.pop()
                
search([])
```

```python

# こっちの方がわかりやすくて好き
count = 0
board = []

def deplication(x, y):
    for y1 in range(y):
        x1 = board[y1]
         # 斜め位置ということは正方形を模すので置いてあるクイーンと置きたいクイーンのxy座標の差の絶対値が等しくなる
        if abs(y - y1) == abs(x - x1):
            return True
    return False

def queen(n, y):
    """
    y(列)をindex
    x(行)をvalu
    """
    global count
    if n == y:
        print(board)
        count += 1
    else:
        for x in range(n):
            if x in board: # 同じ行に値の重複はないため
                continue
            if deplication(x, y): # 斜め方向のチェック
                continue
            board.append(x)
            queen(n, y + 1)
            board.pop()

queen(8, 0)
print(count)
```
